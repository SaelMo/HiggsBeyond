```python
import ROOT
from ROOT import *
import numpy as np
```

#### Reading in ROOT files

```python
SM_Root = TFile("ROOTfiles/GF_HHH_04.root")
TRSM_Root = TFile("ROOTfiles/TRSM_pp_hhh.root")
```

#### Get the tree from the files

```python
tree_SM = SM_Root.Get("LHEF");
tree_TRSM = TRSM_Root.Get("LHEF");
```

#### Transverse momentum
---

```python
c = TCanvas()
h1 = TH1F("h1", "PT", 500, 0, 1000)

h2 = TH1F("h2", "PT", 500, 0, 1000)
tree_TRSM.Project("h2","Particle.PT", "Particle.PID==25")
h2.Draw("E")
h2.SetLineColor(ROOT.kRed)
h2.Scale(1/h2.Integral())


tree_SM.Project("h1","Particle.PT", "Particle.PID==25")
h1.Draw("E same")
h1.SetLineColor(ROOT.kGreen)
h1.Scale(1/h1.Integral())


legend = TLegend()
legend.SetTextFont(72)
legend.SetTextSize(0.04)
legend.AddEntry(h1, 'SM', 'LE')
legend.AddEntry(h2, 'TRSM', 'LE')
legend.Draw('Same')

#c.SetLogy()
c.Draw()
```
